# Data Preprocessing

# Dropping a column.

```python
#Column [Loaned from] is droped.
df.drop(labels="Loaned From",axis=1,inplace=True)
df.head()
```

# Removing prefix and suffix and converting it into whole values for value ,wages and release clause.

```python
def amount_p_s(amount):
    try:
    
        value = float(amount[1:-1])
        suffix = amount[-1:]

        if suffix == 'M':
            value = value * 1000000
        elif suffix == 'K':
            value = value * 1000
    except ValueError:
        value = 0
    return value
    

#removing prefix and suffix and converting it into whole values for value ,wages and release clause.
df['Value'] = df['Value'].apply(amount_p_s)
df['Wage'] = df['Wage'].apply(amount_p_s)
df['Release Clause'] = df['Release Clause'].astype(str)
df['Release Clause'] = df['Release Clause'].apply(amount_p_s)
```

```python
df.head()
```

```python
print("dtype of value is:",df.Value.dtype)
print("dtype of wage is:",df.Wage.dtype)
print("dtype of Release Clause is:",df["Release Clause"].dtype)
```

# Converting column Joined to int with only year

```python
df.Joined.dtype
```

```python
def joined_year(year):
    try:
        yr=year.split(",")
        value = int(yr[-1])
        
    
    except ValueError:
        value = 0
    return value

df['Joined'] = df['Joined'].astype(str)
df['Joined'] = df['Joined'].apply(joined_year)    
```

```python
df.head()
```

```python
df.Joined.dtype
```

# Converting column contract valid until to int with only year

```python
df["Contract Valid Until"].dtype
```

```python
def contract_year(year):
    try:
        yr=year.split(",")
        value = int(yr[-1])
        
    
    except ValueError:
        value = 0
    return value

df['Contract Valid Until'] = df['Contract Valid Until'].astype(str)
df['Contract Valid Until'] = df['Contract Valid Until'].apply(contract_year)
```

```python
df["Contract Valid Until"].dtype
```

# Converting Height from Feet to Centimetre.

```python
def height_cm(ht):
    try:
    
        height = ht.split("'")
        feet = int(height[0])
        inch = int(height[1])
        value=round(((12*feet) + inch)*2.54)
    except ValueError:
        value = 0
    return value

df['Height'] = df['Height'].astype(str)
df['Height'] = df['Height'].apply(height_cm)
```

```python
df.Height.dtype
```

# Removing suffix from weight.

```python
df['Weight'] = df['Weight'].astype(str).str.replace('lbs', '').astype(float)
```
